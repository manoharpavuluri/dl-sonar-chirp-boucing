````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
import warnings
warnings.filterwarnings('ignore')
````

````python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data',header=None)
df
````

````python
df.describe(include='all')
````

````python
#will return name of columns that contains null rows - looks like none. good!

df.columns[df.isnull().any()].tolist()
````

````python
# lets see distinct values in the target column ie column 60

df[60].value_counts()

````

* M stands for mineral

* R stands for rock

````python
from re import A
# lets create x and y

X = df.drop(60,axis=1)
y = df[60]

y.head()
````

````python
X.head()
````

````python
# one hot encoding on column 60 to convert R and M to integer
# and since it has only 2 values,  drop 1 and it will automatically assign the other 0

y = pd.get_dummies(y,drop_first=True)
y.sample(10)


````

````python
y.value_counts()
````

````python
df[60].value_counts()
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=1)
````

````python
X_train.shape, X_test.shape
````

````python
import tensorflow as tf
from tensorflow import keras
````

````python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)
````

````python
model.evaluate(X_test, y_test)
````

````python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print('predict : ', y_pred[:10])
print('test : ', y_test[:10].transpose())
````

````python
y_test[:10]
````

````python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
````



```
## Model with Dropout Layer
```



````python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8)
````

````python
modeld.evaluate(X_test, y_test)
````

````python
y_pred = modeld.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
````

````python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
````

You can see that by using dropout layer test accuracy increased from 0.77 to 0.81
